## Traitment of missing values

In this data we have only the qualitatif variable, so wre use the technique of most frequent.

```python
from sklearn.preprocessing import Imputer
from sklearn.impute import SimpleImputer
import numpy as np
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df2=imp.fit_transform(d_quest)
```

```python
df2[1]
```

```python
dff2=pd.DataFrame(df2)
```

```python
dff2.head()
```

```python
principalDf = pd.DataFrame(data = df2
             , columns = ['SEQN', 'DIQ010', 'DIQ160', 'DIQ170', 'DIQ172',  'DIQ180', 'DIQ050'])
```
